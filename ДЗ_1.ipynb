```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://ru.wikipedia.org/wiki/Трое_в_лодке,_не_считая_собаки'
r_wiki = requests.get(url=url)
```

```python
r_wiki
```

```python
r_wiki.text
```

```python
soup_wiki = BeautifulSoup(r_wiki.text, 'html.parser')
soup_wiki
```

нужен парсинг


```python
pars_wiki = soup_wiki.find_all('p')
pars_wiki
```

```python
texts_wiki = [text.get_text() for text in pars_wiki]
texts_wiki
```

```python
texts_wiki[1]
```

Далее я попробовала отделить знаки пунктуации от слов, чтобы создать словарик частотных слов, это было сделано в рамках интереса, дополнительно.

```python
sp = '!"#$%&()*+,./:;<=>?@[\]^_`{|}~«»„“...—1234567890'
new_wiki = texts_wiki[1]
for item in sp:
  if item in new_wiki:
    new_wiki = new_wiki.replace(item, '')
new_wiki.strip()
```

Далее я создаю пустой словарь для подсчета частотности слов отдельного предложения


```python
count = {}
for item in new_wiki.split():

    if item in count:
        count[item] += 1
    else:
        count[item] = 1
print(count)
```

Это было для каждого отдельного предложения, теперь посмотрим на частотный словарь всего текста:

```python
def all_symb(symb):
  result = ''
  for s in symb:
    result += '' + s
  return result
```

```python
all_wiki = all_symb(texts_wiki)
all_wiki
```

```python
sp = '!"#$%&()*+,./:;<=>?@[\]^_`{|}~«»„“...—1234567890'
for item in sp:
  if item in all_wiki:
    all_wiki = all_wiki.replace(item,'')
all_wiki
```

```python
count = {}
for item in all_wiki.split():

    if item in count:
        count[item] += 1
    else:
        count[item] = 1
print(count)
```
